```python
route_efficiency = receiving_summary[['player', 'team_name', 'routes', 'yards']].groupby(by='team_name', as_index=False).sum()
route_efficiency['avg_yards_per_route'] = route_efficiency['yards'] / route_efficiency['routes']
route_efficiency
```

```python
plt.figure(figsize=(15, 5))
sb.barplot(data=route_efficiency.sort_values(by='avg_yards_per_route', ascending=False),
           x='team_name', y='avg_yards_per_route')
```

```python
concept_routes = receiving_scheme[['player', 'team_name', 'man_yards',
                                   'man_routes', 'zone_yards', 'zone_routes']].groupby(by='team_name', as_index=False).sum()
concept_routes['avg_man_yards_per_route'] = concept_routes['man_yards'] / concept_routes['man_routes']
concept_routes['avg_zone_yards_per_route'] = concept_routes['zone_yards'] / concept_routes['zone_routes']
concept_routes
```

```python
concept_routes_melt = concept_routes.melt(id_vars=['team_name'],
                                          value_vars=['avg_man_yards_per_route', 'avg_zone_yards_per_route'],
                                          var_name='scheme', value_name='avg_yards_per_route')
concept_routes_melt
```

```python
concept_routes_melt.replace(to_replace={'avg_man_yards_per_route': 'man', 'avg_zone_yards_per_route': 'zone'},
                            inplace=True)
concept_routes_melt.sort_values(by=['avg_yards_per_route', 'scheme'], inplace=True, ascending=False)
concept_routes_melt
```

```python
plt.figure(figsize=(15, 5))
sb.barplot(data=concept_routes_melt,
           x='team_name', y='avg_yards_per_route', hue='scheme')
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15, 10))
sb.barplot(data=concept_routes.sort_values(by='avg_man_yards_per_route', ascending=False),
           x='team_name', y='avg_man_yards_per_route', ax=ax[0])
sb.barplot(data=concept_routes.sort_values(by='avg_zone_yards_per_route', ascending=False),
           x='team_name', y='avg_zone_yards_per_route', ax=ax[1])
```

```python
concept_routes['avg_yprr_perc_diff'] = (concept_routes['avg_man_yards_per_route'] - concept_routes['avg_zone_yards_per_route']) / concept_routes['avg_zone_yards_per_route'] * 100
```

```python
plt.figure(figsize=(15,5))
ax = sb.barplot(data=concept_routes.sort_values(by='avg_yprr_perc_diff', ascending=False),
           x='team_name', y='avg_yprr_perc_diff')
ax.set(xlabel = 'Team Name', ylabel = 'YPRR Percent Difference Between Man and Zone')
```

```python
summary_subset = receiving_summary.sort_values(by='yards', ascending=False)[['player', 'player_id', 'position', 'team_name', 'avg_depth_of_target', 'receptions', 'yards']]
summary_subset
```

```python
summary_subset = pd.merge(summary_subset,
                          receiving_depth[['player_id', 'deep_avg_depth_of_target', 'deep_receptions', 'deep_yards']],
                          on='player_id')
summary_subset
```

```python
summary_subset['deep_recs_perc'] = summary_subset['deep_receptions'] / summary_subset['receptions'] * 100
summary_subset['deep_yards_perc'] = summary_subset['deep_yards'] / summary_subset['yards'] * 100
summary_subset['avg_depth_ratio'] = summary_subset['deep_avg_depth_of_target'] / summary_subset['avg_depth_of_target']
```

```python
summary_subset
```

```python
sb.scatterplot(data=summary_subset, x='deep_recs_perc', y='deep_yards_perc', hue='position')
```

```python
summary_subset['deep_pass_efficiency'] = summary_subset['deep_yards_perc'] / summary_subset['deep_recs_perc']
summary_subset
```

```python
plt.figure(figsize=(15, 10))
sb.barplot(data=summary_subset.sort_values(by='deep_pass_efficiency', ascending=False).head(20), x='player', y='deep_pass_efficiency')
plt.xticks(rotation=70);
```

```python
summary_subset.sort_values(by='deep_pass_efficiency', ascending=False).head(20)['receptions']
```

```python
plt.figure(figsize=(15, 10))
sb.barplot(data=summary_subset.sort_values(by='yards', ascending=False).head(20), x='player', y='deep_pass_efficiency')
plt.xticks(rotation=70);
```

```python
sb.scatterplot(data=summary_subset[(summary_subset['avg_depth_ratio'] >= 0) & (summary_subset['avg_depth_ratio'] < 10)],
               x='avg_depth_ratio', y='deep_pass_efficiency', hue='position')
```

```python
sb.scatterplot(data=summary_subset[(summary_subset['avg_depth_ratio'] >= 0)
                                   & (summary_subset['avg_depth_ratio'] < 5)
                                   & (summary_subset['deep_pass_efficiency'] <= 5)],
               x='avg_depth_ratio', y='deep_pass_efficiency', hue='position')
```
